```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

from neuralforecast import NeuralForecast
from neuralforecast.models import DLinear
from neuralforecast.utils import AirPassengersDF


train_df = pd.read_csv('train.csv.zip')
test_df = pd.read_csv('test.csv.zip')
stores_df = pd.read_csv('stores.csv')
features_df = pd.read_csv('features.csv.zip')
print("Loaded actual Walmart data from CSV files")

print(f"Train data shape: {train_df.shape}")
print(f"Test data shape: {test_df.shape}")
```

```python
def preprocess_walmart_data(train_df, test_df, stores_df):

    train_df['Date'] = pd.to_datetime(train_df['Date'])
    test_df['Date'] = pd.to_datetime(test_df['Date'])

    train_df = train_df.merge(stores_df, on='Store', how='left')
    test_df = test_df.merge(stores_df, on='Store', how='left')

    train_df['unique_id'] = train_df['Store'].astype(str) + '_' + train_df['Dept'].astype(str)
    test_df['unique_id'] = test_df['Store'].astype(str) + '_' + test_df['Dept'].astype(str)

    train_df = train_df.sort_values(['unique_id', 'Date'])
    test_df = test_df.sort_values(['unique_id', 'Date'])

    return train_df, test_df
```

```python
train_df, test_df = preprocess_walmart_data(train_df, test_df, stores_df)
```

```python
def prepare_neural_forecast_data(train_df, test_df):

    train_nf = train_df[['unique_id', 'Date', 'Weekly_Sales']].copy()
    train_nf.columns = ['unique_id', 'ds', 'y']

    train_nf = train_nf.dropna()

    forecast_horizon = len(test_df['Date'].unique())

    return train_nf, forecast_horizon
```

```python
def calculate_wmae(y_true, y_pred, is_holiday):

    weights = np.where(is_holiday, 5, 1)

    weighted_errors = weights * np.abs(y_true - y_pred)

    wmae = np.sum(weighted_errors) / np.sum(weights)

    return wmae
```

```python
def evaluate_model(train_df, full_train_df, args, forecast_horizon=25):

    train_df_sorted = train_df.sort_values('ds')
    cutoff_date = train_df_sorted['ds'].max() - pd.Timedelta(weeks=forecast_horizon)

    train_val = train_df_sorted[train_df_sorted['ds'] <= cutoff_date]
    test_val = train_df_sorted[train_df_sorted['ds'] > cutoff_date]

    # print(f"shape of train_val - {train_val.shape}")
    # print(f"shape of test_val - {test_val.shape}")

    nf_val = NeuralForecast(
        models=[DLinear(h=forecast_horizon, learning_rate=args['lr'], input_size=args['input_size'], max_steps=args['steps'])],
        freq='W-FRI'
    )

    nf_val.fit(train_val)
    forecasts_val = nf_val.predict()

    # print(forecasts_val)
    # print(test_val.sort_values(['unique_id', 'ds']))

    comparison = test_val.merge(
        forecasts_val,
        on=['unique_id', 'ds'],
        how='inner'
    )

    # print(comparison.head(10))


    # print(train_df)

    comparison = comparison.merge(
        full_train_df[['unique_id', 'ds', 'IsHoliday']].drop_duplicates(),
        on=['unique_id', 'ds'],
        how='left'
    )

    comparison['IsHoliday'] = comparison['IsHoliday'].fillna(False)

    mae = np.mean(np.abs(comparison['y'] - comparison['DLinear']))
    wmae = calculate_wmae(comparison['y'], comparison['DLinear'], comparison['IsHoliday'])
    mape = np.mean(np.abs((comparison['y'] - comparison['DLinear']) / comparison['y'])) * 100

    print(f"\nValidation Result:")
    print(f"WMAE: {wmae:.2f}")

    holiday_mask = comparison['IsHoliday']
    if holiday_mask.sum() > 0:
        holiday_mae = np.mean(np.abs(comparison[holiday_mask]['y'] - comparison[holiday_mask]['DLinear']))
        non_holiday_mae = np.mean(np.abs(comparison[~holiday_mask]['y'] - comparison[~holiday_mask]['DLinear']))
        print(f"Holiday MAE: {holiday_mae:.2f}")
        print(f"Non-Holiday MAE: {non_holiday_mae:.2f}")

    return wmae
```

```python
def create_submission(forecasts, test_df):

    submission = test_df.merge(
        forecasts[['unique_id', 'Date', 'DLinear']].drop_duplicates(),
        on=['unique_id', 'Date'],

        how='left'
    )


    submission['DLinear'] = submission['DLinear'].fillna(submission['DLinear'].median())

    submission_df = pd.DataFrame({
        'Id': submission['Store'].astype(str) + '_' +
              submission['Dept'].astype(str) + '_' +
              submission['Date'].dt.strftime('%Y-%m-%d'),
        'Weekly_Sales': submission['DLinear']
    })

    return submission_df
```

```python
param_grid = {
    'input_size': [26, 52, 78],
    'learning_rate': [1e-3, 5e-4],
    'max_steps': [50, 100],
}


import itertools

best_wmae = float('inf')
best_params = None
best_model = None

for input_size, learning_rate, max_steps in itertools.product(
        param_grid['input_size'],
        param_grid['learning_rate'],
        param_grid['max_steps']
):
    print(f"\nTrying input_size={input_size}, learning_rate={learning_rate}, "
          f"max_steps={max_steps}")

    try:
        wmae = evaluate_model(
            train_df=train_nf_sample,
            full_train_df=train_df_sorted,
            args={'lr': learning_rate, "input_size": input_size,
                  "steps":max_steps}
        )

        print(f" WMAE: {wmae:.2f}")

        if wmae < best_wmae:
            best_wmae = wmae
            best_params = {
                'input_size': input_size,
                'learning_rate': learning_rate,
                'max_steps': max_steps,
            }

    except Exception as e:
        print(f" Failed with error: {str(e)}")

print(f"best_wmae - {best_wmae}")
print(f"best_params_overall - {best_params}")
```
