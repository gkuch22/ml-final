```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error


train = pd.read_csv("/kaggle/input/walmart-recruiting-store-sales-forecasting/train.csv.zip", parse_dates=["Date"])
features = pd.read_csv("/kaggle/input/walmart-recruiting-store-sales-forecasting/features.csv.zip", parse_dates=["Date"])
stores = pd.read_csv("/kaggle/input/walmart-recruiting-store-sales-forecasting/stores.csv")
```

```python
full = train.merge(features, on=["Store", "Date"], how="left")
full = full.merge(stores, on="Store", how="left")
```

```python
markdown_cols = ["MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5"]
for col in markdown_cols:
    if col in full.columns:
        full[col] = full[col].fillna(0)
```

```python
full["Type"] = full["Type"].astype("category")
```

```python
full = full.sort_values(by=["Store", "Dept", "Date"]).reset_index(drop=True)
```

```python
full
```

```python
full["year"] = full["Date"].dt.year
full["month"] = full["Date"].dt.month
full["week"] = full["Date"].dt.isocalendar().week.astype(int)
full["dayofweek"] = full["Date"].dt.dayofweek
full["is_weekend"] = full["dayofweek"] >= 5
```

```python
full.isna().mean()
```

```python
def create_lag_features(data, lags=[1, 52]):
    for lag in lags:
        data[f"lag_{lag}"] = data.groupby(["Store", "Dept"])["Weekly_Sales"].shift(lag)
    return data
```

```python
def create_rolling_features(data, windows=[3]):
    for window in windows:
        data[f"rolling_mean_{window}"] = (
            data.groupby(["Store", "Dept"])["Weekly_Sales"]
            .shift(1)
            .rolling(window)
            .mean()
        )
    return data
```

```python
full = create_lag_features(full)
full = create_rolling_features(full)

full = full.dropna()
```

```python
full.isna().mean()
```

```python
train_df = full[full["Date"] <= "2012-05-31"]
valid_df = full[(full["Date"] > "2012-05-31") & (full["Date"] <= "2012-08-31")]
test_df = full[(full["Date"] > "2012-08-31")]

print("Train rows:", len(train_df))
print("Valid rows:", len(valid_df))
print("Test rows:", len(test_df))
```

```python
train_df.head()
```

```python
target = "Weekly_Sales"
categorical = ["Store", "Dept", "Type", "IsHoliday_x"]
features = [
    "Store", "Dept", "Type", "IsHoliday_x", "Size", "Temperature", "Fuel_Price",
    "CPI", "Unemployment", "year", "month", "week", "dayofweek", "is_weekend",
    "lag_1", "lag_52", "rolling_mean_3"
]

lgb_train = lgb.Dataset(train_df[features], train_df[target], categorical_feature=categorical, free_raw_data=False)
lgb_valid = lgb.Dataset(valid_df[features], valid_df[target], categorical_feature=categorical, free_raw_data=False)
```

```python
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'seed': 42,
    'verbose': -1
}
```

```python
def WMAE(y_true, y_pred, is_holiday):
    weights = np.where(is_holiday, 5, 1)
    return np.sum(weights * np.abs(y_true - y_pred)) / np.sum(weights)
```

```python
param_grid = {
    "learning_rate": [0.03, 0.05, 0.1],
    "num_leaves": [31, 63, 127],
    "feature_fraction": [0.8, 1.0],
    "bagging_fraction": [0.8, 1.0],
}

best_wmae = float("inf")
best_params = None

for lr in param_grid["learning_rate"]:
    for leaves in param_grid["num_leaves"]:
        for feat_frac in param_grid["feature_fraction"]:
            for bag_frac in param_grid["bagging_fraction"]:

                params = {
                    'objective': 'regression',
                    'metric': 'rmse',
                    'boosting_type': 'gbdt',
                    'learning_rate': lr,
                    'num_leaves': leaves,
                    'feature_fraction': feat_frac,
                    'bagging_fraction': bag_frac,
                    'bagging_freq': 5,
                    'seed': 42,
                    'verbose': -1
                }

                model = lgb.train(
                    params,
                    lgb_train,
                    valid_sets=[lgb_train, lgb_valid],
                    num_boost_round=2000
                )

                val_preds = model.predict(valid_df[features])
                val_score = WMAE(valid_df[target].values, val_preds, valid_df["IsHoliday_x"].values)

                print(f"lr={lr}, leaves={leaves}, feat_frac={feat_frac}, bag_frac={bag_frac} -> WMAE: {val_score:.4f}")

                if val_score < best_wmae:
                    best_wmae = val_score
                    best_params = params

print("Best WMAE:", best_wmae)
print("Best Params:", best_params)

model = lgb.train(
    best_params,
    lgb_train,
    valid_sets=[lgb_train, lgb_valid],
    num_boost_round=2000,
)
```

```python
train_preds = model.predict(train_df[features])
val_preds = model.predict(valid_df[features])
test_preds = model.predict(test_df[features])

def WMAE(y_true, y_pred, is_holiday):
    weights = np.where(is_holiday, 5, 1)
    return np.sum(weights * np.abs(y_true - y_pred)) / np.sum(weights)

train_score = WMAE(train_df[target].values, train_preds, train_df["IsHoliday_x"].values)
val_score = WMAE(valid_df[target].values, val_preds, valid_df["IsHoliday_x"].values)
test_score = WMAE(test_df[target].values, test_preds, test_df["IsHoliday_x"].values)

print(f"Train WMAE: {train_score:.4f}")
print(f"Validation WMAE: {val_score:.4f}")
print(f"Test WMAE: {test_score:.4f}")
```
